```python
# Step b: Create 4 variables

# str
gym_member = "Alex Alliton"
# float
preferred_weight_kg = 20.5
# int
highest_reps = 25
# bool
membership_active = True
```

```python
# Step c: Create a dictionary named workout_stats

# dict[str, tuple[int, int, int]]
workout_stats = {
    "Alex": (30, 45, 20),
    "Jamie": (40, 30, 50),
    "Taylor": (25, 35, 40),
    "Jordan": (50, 40, 35)
}
```

```python
# Step d: Calculate total workout minutes using a loop and add to dictionary

for friend in list(workout_stats.keys()):
    workout_stats[friend + "_Total"] = sum(workout_stats[friend])

```

```python
# Step e: Create a 2D nested list called workout_list

# list[list[int]]
friends = [name for name in workout_stats if "_Total" not in name]
workout_list = [list(workout_stats[name]) for name in friends]
workout_list
```

```python
# Step f: Slice the workout_list

print("Yoga & Running (all friends):")
for row in workout_list:
    print(row[0:2])

print("Weightlifting (last two friends):")
for row in workout_list[-2:]:
    print(row[2])
```

```python
# Step g: Check if any friend's total >= 120

for name in friends:
    total = workout_stats[f"{name}_Total"]
    if total >= 120:
        print(f"Great job staying active, {name}!")
```

```python
# Step h: User input to look up a friend

search_name = input("Enter a friend's name: ").strip()

if search_name in workout_stats:
    yoga, running, weightlifting = workout_stats[search_name]
    total = workout_stats[f"{search_name}_Total"]
    print(f"{search_name} workout minutes:")
    print("Yoga:", yoga)
    print("Running:", running)
    print("Weightlifting:", weightlifting)
    print("Total:", total)
else:
    print(f"Friend {search_name} not found in the records.")

```

```python
# Step i: Friend with highest and lowest total workout minutes

totals = {name: workout_stats[f"{name}_Total"] for name in friends}
highest_friend = max(totals, key=totals.get)
lowest_friend = min(totals, key=totals.get)

print("Highest total:", highest_friend, totals[highest_friend])
print("Lowest total:", lowest_friend, totals[lowest_friend])
```
